## Mexclando todos los conceptos aprendidos

```python
proyecto = input("Escribe el nombre de tu proyecto: ")
horas_estimadas = input("Escribe las horas que vas a dedicar a tu proyecto: ")
moneda_seleccionada = input("Escribe la moneda que deseas usar (COP, USD, EUR): ")
valor_hora = input("¿Cuál es el valor de la hora trabajada?: ")
plazo_estimado = input("¿Cuál es el plzao estimado para el proyecto?: ")
```

```python
print(proyecto)
print(horas_estimadas)
print(valor_hora)
print(plazo_estimado)
```

## Cálculos finales del proyecto

```python
valor_total = float(horas_estimadas) * float(valor_hora)

simbolo_moneda = ""

if moneda_seleccionada == "COP":
    simbolo_moneda = "$"
elif moneda_seleccionada == "USD":
    simbolo_moneda = "$"
elif moneda_seleccionada == "EUR":
    simbolo_moneda = "€"

print(f"El valor final de tu proyecto es: {simbolo_moneda} {valor_total:.0f} {moneda_seleccionada}")
```

## Generando archivos PDF del presupuesto

* Crear el archvio PDF
* Definir el tipo de letra
* Llenar el documento
* Guardar el entregable

### Librería necesaria

* `pip install fpdf`


```python
from fpdf import FPDF
```

```python

# ------------------------------
# Cambio del tamaño de la página
# ------------------------------
# Opción 1: Usar un formato estándar (A5, Letter, etc.)
# pdf = FPDF(orientation='P', unit='mm', format='Letter')
# Opción 2: Definir un tamaño personalizado (ej. 100x150 mm)
pdf = FPDF(orientation='P', unit='mm', format=(215.9, 279.4))
# pdf = FPDF()

pdf.add_page()
pdf.set_font("Arial")

# Insertar imagen de fondo (x, y, ancho, alto)
# w=210 y h=297 es tamaño A4 vertical
# w=215.9 y h=279.4 es tamaño Letter
# pdf.image('ruta/a/tu/imagen.jpg', x=0, y=0, w=215.9, h=279.4)
# pdf.image("template.png", x=0, y=0)
pdf.image("template-letter-presupuesto.png", x=0, y=0, w=215.9, h=279.4)

# Escribir el texto
pdf.text(118, 153, proyecto)
pdf.text(118, 168, horas_estimadas)
pdf.text(118, 184, f"$ {float(valor_hora):.2f}")
pdf.text(118, 198, plazo_estimado)

pdf.set_text_color(255, 255, 255)
pdf.set_font_size(18)
pdf.text(118, 215, f"$ {float(valor_total):.2f}")

# Crear el archivo pdf
pdf.output("presupuesto.pdf")
```
